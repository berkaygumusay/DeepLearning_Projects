```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
#Loading The data

raw_data = pd.read_csv('weatherAUS.csv')
raw_data['RainToday'] = raw_data['RainToday'].map({'Yes':1, 'No':0})
raw_data['RainTomorrow'] = raw_data['RainTomorrow'].map({'Yes':1, 'No':0})
raw_data.head()
```

```python
raw_data.columns
```

```python
#Picking The Data of Sydney

sydney_raw_data = raw_data[raw_data['Location'] == 'Sydney']
sydney_raw_data = sydney_raw_data.drop(['Date','Location','WindGustDir','WindDir9am','WindDir3pm'],axis=1)
sydney_raw_data = sydney_raw_data.dropna(thresh = len(sydney_raw_data)*0.8,axis=1)
sydney_raw_data = sydney_raw_data.dropna()
sydney_raw_data
```

```python
from sklearn.preprocessing import MinMaxScaler
#Scaling The Data Between 0 and 1

scaler = MinMaxScaler()
sydney_raw_data = pd.DataFrame(scaler.fit_transform(sydney_raw_data), columns=sydney_raw_data.columns)
sydney_raw_data.head()
```

```python
from sklearn.model_selection import train_test_split

#Splitting The Train & Test Sets

x = sydney_raw_data.drop(['RainTomorrow'],axis=1)
y = sydney_raw_data['RainTomorrow']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=11)
```

```python
#Creating The Model

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(16,activation='relu',input_shape=(len(x_train.columns),)),
    tf.keras.layers.Dense(32,activation='relu'),
    tf.keras.layers.Dense(64,activation='relu'),
    tf.keras.layers.Dense(32,activation='relu'),
    tf.keras.layers.Dense(16,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
   # tf.keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.fit(x_train,y_train,epochs=50)
```

```python
#Testing The Model

model.evaluate(x_test,y_test)
```

```python
#Saving The Model
model.save('weather_model.h5')

#Loading The Model
model = tf.keras.models.load_model('weather_model.h5')

predictions = model.predict(x_test)
predictions = np.round(predictions)
predictions[:10]
```
